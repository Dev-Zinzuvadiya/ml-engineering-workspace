```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

stored_Testing_Model = pd.read_csv("test_dataset.csv")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

print(f">> TOTAL SHAPE OF 'TESTING_MODEL' CSV FILE : {stored_Testing_Model.shape}")
```

```python
print("-->> DISPLAY 'TESTING_MODEL' DATASET ↘")
display(stored_Testing_Model.head())

X_train_STV = stored_Testing_Model
```

```python
NumColsSTV = X_train_STV.select_dtypes(include=['int64', 'float64']).columns
CatColsSTV = X_train_STV.select_dtypes(include='object').columns

X_train_STV[NumColsSTV].isnull().sum()
```

```python
impute_mean = SimpleImputer(strategy='mean')
display(impute_mean.fit(X_train_STV[NumColsSTV]))

print("\n", impute_mean.statistics_)
```

```python
filled_values = impute_mean.transform(X_train_STV[NumColsSTV])
X_train_STV[NumColsSTV] = filled_values
```

```python
impute_constant = SimpleImputer(strategy='constant', fill_value=90)
X_train_STV[CatColsSTV].isnull().sum()
```

```python
impute_mode = SimpleImputer(strategy='most_frequent')
impute_mode.fit(X_train_STV[CatColsSTV])
impute_mode
```

```python
X_train_STV[CatColsSTV] = impute_mode.transform(X_train_STV[CatColsSTV])
X_train_STV[CatColsSTV].isnull().sum()
```

```python
total_missing = X_train_STV.isnull().sum().sum()
print(f">> TOTAL MISSING VALUES IN ENTIRE DATASET : {total_missing}")
```
