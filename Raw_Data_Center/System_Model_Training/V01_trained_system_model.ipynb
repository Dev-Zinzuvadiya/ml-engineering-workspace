```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

stored_Trained_Model = pd.read_csv("trained_dataset.csv")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

print(f">> TOTAL SHAPE OF 'TRAINED_MODEL' CSV FILE : {stored_Trained_Model.shape}")
```

```python
print("-->> DISPLAY 'TRAINED_MODEL' DATASET ↘")
display(stored_Trained_Model.head())

X_train_STT = stored_Trained_Model.drop('SalePrice', axis=1)
Y_train_STT = stored_Trained_Model['SalePrice']
```

```python
print(f">> REMOVE 'SalePrice' COLUMN FROM 'TRAINED_MODEL' {X_train_STT.shape}")
print(f">> ONLY 'SalePrice' COLUMN SHOWS FROM 'TRAINED_MODEL' {Y_train_STT.shape}\n\n")

NumColsSTT = X_train_STT.select_dtypes(include=['int64', 'float64']).columns
CatColsSTT = X_train_STT.select_dtypes(include='object').columns

X_train_STT[NumColsSTT].isnull().sum()
```

```python
impute_mean = SimpleImputer(strategy='mean')
display(impute_mean.fit(X_train_STT[NumColsSTT]))

print("\n", impute_mean.statistics_)
```

```python
filled_values = impute_mean.transform(X_train_STT[NumColsSTT])
X_train_STT[NumColsSTT] = filled_values
```

```python
impute_constant = SimpleImputer(strategy='constant', fill_value=90)
X_train_STT[CatColsSTT].isnull().sum()
```

```python
impute_mode = SimpleImputer(strategy='most_frequent')
impute_mode.fit(X_train_STT[CatColsSTT])
impute_mode
```

```python
X_train_STT[CatColsSTT] = impute_mode.transform(X_train_STT[CatColsSTT])
X_train_STT[CatColsSTT].isnull().sum()
```

```python
total_missing = X_train_STT.isnull().sum().sum()
print(f">> TOTAL MISSING VALUES IN ENTIRE DATASET : {total_missing}")
```
