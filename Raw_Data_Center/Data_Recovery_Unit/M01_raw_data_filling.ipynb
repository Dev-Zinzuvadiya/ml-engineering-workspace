```python
# GETHER DATASET & IMPOER LIBRARY
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('trained_dataset.csv')

pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)

display(df.shape)
display(df.head(5))
```

```python
null_values_in_perc = df.isnull().sum() / df.shape[0] * 100
drp_cols = null_values_in_perc[null_values_in_perc > 20].keys()
drpd_cols = df.drop(columns=drp_cols)
df_vals_dtype = drpd_cols.select_dtypes(include=['int64', 'float64'])

cols_names=[i for i in df_vals_dtype.columns if df_vals_dtype[i].isnull().sum() > 0]
```

```python
plt.figure(figsize=(10,10))
sns.set()

for i,j in enumerate(cols_names):
    plt.subplot(2,2,i+1)
    # sns.distplot(df_vals_dtype[j], bins=20, kde_kws={'linewidth':5,'color':'red'})
    sns.distplot(df_vals_dtype[j], bins=20, kde=True)
```

```python
df_af_mean = df_vals_dtype.fillna(df_vals_dtype.mean())
display(df_af_mean.isnull().sum().sum())
display(df_af_mean)
```

```python
plt.figure(figsize=(10,10))
sns.set()

for i,j in enumerate(cols_names):
    plt.subplot(2,2,i+1)
    # sns.distplot(df_vals_dtype[j], bins=20, kde_kws={'linewidth': 8,'color': 'red'}, label='original')
    sns.histplot(df_vals_dtype[j], bins=20, kde=True, label='original')
    
    # sns.distplot(df_af_mean[j], bins=20, kde_kws={'linewidth': 5,'color': 'red'}, label='new')
    sns.histplot(df_af_mean[j], bins=20, kde=True, label='new')
    plt.legend()
```

```python
df5 = df_vals_dtype.fillna(df_vals_dtype.median())
display(df5.isnull().sum().sum())

plt.figure(figsize=(10,10))
sns.set()

for i,j in enumerate(cols_names):
    plt.subplot(2,2,i+1)
    sns.distplot(df_vals_dtype[j], bins=20, hist=False, kde_kws={'linewidth':4,'color':'red'},label='original')
    sns.distplot(df_af_mean[j], bins=20, hist=False, kde_kws={'linewidth':4,'color':'green'},label='mean')
    sns.distplot(df5[j], bins=20, hist=False, kde_kws={'linewidth':4,'color':'blue'},label='median')
    plt.legend()
```

```python
df_concat = pd.concat([df_vals_dtype[cols_names], df_af_mean[cols_names], df5[cols_names]], axis=1)
df_concat[df_concat.isnull().any(axis=1)]
```
