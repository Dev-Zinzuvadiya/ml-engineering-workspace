```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('trained_dataset.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
print(f">> TOTAL ROWS AND COLUMNS SIZEs : '{df.shape}'")
```

```python
cat_data = df.select_dtypes(include='object')
print(f">> TOTAL ROWS AND COLUMNS SIZEs (ONLY OBEJECTs) : '{cat_data.shape}'")
cat_data.isnull().sum()
```

```python
# cat_data.isnull().mean() * 100
per_cat = cat_data.isnull().sum() / df.shape[0] * 100
col = per_cat[per_cat > 20].keys()
cat_data.drop(columns=col, axis=1, inplace=True)

print(f"DATASET SHAPE AFTER REMOVE NULL VALUEs : '{cat_data.shape}'")

cat_data.isnull().mean() * 100
```

```python
cat_data['BsmtQual'].fillna('Missing')
```
