## >> Import Libraries & load Dataset

```python
# IMPORT THE REQUIRED LIBRARIES FOR DATA
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# READ THE TRAINING AND TEST DATASETS FROM CSV FILES
train_DF = pd.read_csv("train.csv")
test_DF = pd.read_csv("test.csv")

# DISPLAY ALL ROWS AND COLUMNS WITHOUT TRUNCATION
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

# DISPLAY THE DIMENSIONS OF BOTH DATASETS
print(f":> TRAINING DATASET SHAPE: {train_DF.shape}")
print(f":> TEST DATASET SHAPE: {test_DF.shape}")
```

```python
# PREVIEW THE FIRST 15 ROWS OF THE TRAINING DATASET
print("=>> TRAINING DATASET PREVIEW (FIRST 15 ROWS) >>")
display(train_DF.head(15))

# PREVIEW THE FIRST 15 ROWS OF THE TEST DATASET
print("\n=>> TEST DATASET PREVIEW (FIRST 15 ROWS) >>")
display(test_DF.head(15))
```

```python
# COUNT AND DISPLAY MISSING VALUES IN BOTH DATASETS
print("\n=>> MISSING VALUE COUNT IN TRAINING DATASET >>")
display(train_DF.isnull().sum())

print("\n=>> MISSING VALUE COUNT IN TEST DATASET >>")
display(test_DF.isnull().sum())
```

## >> Feature and Target Separation

```python
# SEPARATE FEATURES (X) AND TARGET VARIABLE (Y) FROM TRAINING DATA
X_train = train_DF.drop('SalePrice', axis=1)
Y_train = train_DF['SalePrice']

# DISPLAY THE SHAPE OF FEATURES AND TARGET
print(f":> FEATURES DATASET SHAPE: {X_train.shape}")
print(f":> TARGET VARIABLE SHAPE: {Y_train.shape}")
```

```python
# IDENTIFY NUMERICAL AND CATEGORICAL COLUMNS IN THE FEATURES DATASET
Num_Cols = X_train.select_dtypes(include=['int64', 'float64']).columns
Cate_Cols = X_train.select_dtypes(include=['object']).columns

# DISPLAY MISSING VALUE COUNTS FOR NUMERICAL COLUMNS
print("=>> MISSING VALUES IN NUMERICAL COLUMNS (TRAINING DATA) >>")
display(X_train[Num_Cols].isnull().sum())

# DISPLAY MISSING VALUE COUNTS FOR CATEGORICAL COLUMNS
print("\n=>> MISSING VALUES IN CATEGORICAL COLUMNS (TRAINING DATA) >>")
display(X_train[Cate_Cols].isnull().sum())
```

## >> Imputation Strategy Initialization

```python
# INITIALIZE IMPUTATION STRATEGIES
Mean_Impute = SimpleImputer(strategy='mean')                                # FILL NUMERICAL MISSING VALUES WITH MEAN
Mode_Impute = SimpleImputer(strategy="most_frequent")                       # FILL CATEGORICAL MISSING VALUES WITH MODE
Constant_Impute = SimpleImputer(strategy="constant", fill_value=90)         # FILL WITH CONSTANT VALUE (EXAMPLE)

# FIT THE IMPUTERS TO THE RESPECTIVE DATA
display(Mean_Impute.fit(X_train[Num_Cols]))
display(Mode_Impute.fit(X_train[Cate_Cols]))
```

```python
# VIEW THE CALCULATED STATISTICS (MEAN VALUES FOR NUMERICAL COLUMNS)
Mean_Impute.statistics_
```

```python
# VIEW THE CALCULATED STATISTICS (MODE VALUES FOR CATEGORICAL COLUMNS)
Mode_Impute.statistics_
```

```python
# DISPLAY SAMPLE OF NUMERICAL DATA BEFORE IMPUTATION (TRAINING)
print("=>> NUMERICAL DATA SAMPLE BEFORE IMPUTATION (TRAINING) >>")
display(X_train[Num_Cols].head(15))

# DISPLAY SAMPLE OF CATEGORICAL DATA BEFORE IMPUTATION (TRAINING)
print("\n=>> CATEGORICAL DATA SAMPLE BEFORE IMPUTATION (TRAINING) >>")
display(X_train[Cate_Cols].head(15))
```

```python
# DISPLAY SAMPLE OF NUMERICAL DATA BEFORE IMPUTATION (TEST)
print("=>> NUMERICAL DATA SAMPLE BEFORE IMPUTATION (TEST) >>")
display(test_DF.select_dtypes(include=['int64', 'float64']).head(15))

# DISPLAY SAMPLE OF CATEGORICAL DATA BEFORE IMPUTATION (TEST)
print("\n=>> CATEGORICAL DATA SAMPLE BEFORE IMPUTATION (TEST) >>")
display(test_DF.select_dtypes(include=['object']).head(15))
```

## >> Apply Imputation Strategies

```python
# APPLY MEAN IMPUTATION TO NUMERICAL COLUMNS IN BOTH DATASETS
X_train[Num_Cols] = Mean_Impute.transform(X_train[Num_Cols])
test_DF[Num_Cols] = Mean_Impute.transform(test_DF[Num_Cols])

# VERIFY NO MISSING VALUES REMAIN IN NUMERICAL COLUMNS (TRAINING)
print("=>> MISSING VALUE COUNT IN NUMERICAL COLUMNS AFTER IMPUTATION (TRAINING) >>")
display(X_train[Num_Cols].isnull().sum())

# VERIFY NO MISSING VALUES REMAIN IN NUMERICAL COLUMNS (TEST)
print("\n=>> MISSING VALUE COUNT IN NUMERICAL COLUMNS AFTER IMPUTATION (TEST) >>")
display(test_DF[Num_Cols].isnull().sum())
```

```python
# CHECK FOR ANY REMAINING MISSING VALUES AFTER NUMERICAL IMPUTATION
print(f":> TOTAL MISSING VALUES IN TRAINING DATA AFTER NUMERICAL IMPUTATION: {X_train.isnull().sum().sum()}")
print(f":> TOTAL MISSING VALUES IN TEST DATA AFTER NUMERICAL IMPUTATION: {test_DF.isnull().sum().sum()}")
```

```python
# APPLY MODE IMPUTATION TO CATEGORICAL COLUMNS IN BOTH DATASETS
X_train[Cate_Cols] = Mode_Impute.transform(X_train[Cate_Cols])
test_DF[Cate_Cols] = Mode_Impute.transform(test_DF[Cate_Cols])

# VERIFY NO MISSING VALUES REMAIN IN CATEGORICAL COLUMNS (TRAINING)
print("\n=>> MISSING VALUE COUNT IN CATEGORICAL COLUMNS AFTER IMPUTATION (TRAINING) >>")
display(X_train[Cate_Cols].isnull().sum())

# VERIFY NO MISSING VALUES REMAIN IN CATEGORICAL COLUMNS (TEST)
print("\n=>> MISSING VALUE COUNT IN CATEGORICAL COLUMNS AFTER IMPUTATION (TEST) >>")
display(test_DF[Cate_Cols].isnull().sum())
```

## >> Final Dataset Validation

```python
# FINAL VERIFICATION - CHECK FOR ANY MISSING VALUES IN EITHER DATASET
print(f"\n:> TOTAL MISSING VALUES IN TRAINING DATASET (FINAL): {X_train.isnull().sum().sum()}")
print(f":> TOTAL MISSING VALUES IN TEST DATASET (FINAL): {test_DF.isnull().sum().sum()}")
```
