## >> Encoded Feature Extraction

```python
# RETRIEVE FEATURE NAMES AND CONVERT ENCODED ARRAY TO DATAFRAME
print("=>> RETRIEVING FEATURE NAMES AND CREATING DATAFRAME >>")
feature_names = OneHot_Encode.get_feature_names_out(['sex', 'smoker', 'day', 'time'])
feature_names
```

```python
# CREATE A DATAFRAME FROM THE ENCODED ARRAY WITH PROPER COLUMN NAMES
encoded_DF = pd.DataFrame(encoded_arr, columns=feature_names)
encoded_DF.head(10)
```

## >> Binary Value Analysis Functions

```python
# DEFINE LISTS TO STORE BINARY VALUES FOR COUNTING
One_Binary = []             # STORES VALUES EQUAL TO 1
Zero_Binary = []            # STORES VALUES EQUAL TO 0

# FUNCTION TO SEPARATE BINARY VALUES INTO ONE AND ZERO LISTS
def CHECK_EVERY_TICK_OF_CATEGORY(labels):
    One_Binary.clear()
    Zero_Binary.clear()

    for i in labels:
        if i == 1:
            One_Binary.append(i)
        else:
            Zero_Binary.append(i)


# FUNCTION TO ANALYZE AND DISPLAY COUNTS FOR EACH ENCODED CATEGORY
def CHECK_CATEGORY_BINARY_VALUES(separations):
    if separations == 'sex_Female':
        CHECK_EVERY_TICK_OF_CATEGORY(encoded_DF['sex_Female'])

        print(f":> COUNT OF MALES (0): {Zero_Binary.count(0)}")
        print(f":> COUNT OF FEMALES (1): {One_Binary.count(1)}\n")

    elif separations == 'smoker_No':
        CHECK_EVERY_TICK_OF_CATEGORY(encoded_DF['smoker_No'])

        print(f":> COUNT OF SMOKERS (1): {Zero_Binary.count(0)}")
        print(f":> COUNT OF NON-SMOKERS (0): {One_Binary.count(1)}\n")

    elif separations == 'time_Dinner':
        CHECK_EVERY_TICK_OF_CATEGORY(encoded_DF['time_Dinner'])

        print(f":> TOTAL LUNCH VISITS (0): {Zero_Binary.count(0)}")
        print(f":> TOTAL DINNER VISITS (1): {One_Binary.count(1)}\n")

    elif separations == 'day_Thur' or separations == 'day_Fri' or separations == 'day_Sat' or separations == 'day_Sun':

        if separations == 'day_Thur':
            CHECK_EVERY_TICK_OF_CATEGORY(encoded_DF['day_Thur'])
            print(f":> TOTAL VISITS ON THURSDAY: {One_Binary.count(1)}")

        elif separations == 'day_Fri':
            CHECK_EVERY_TICK_OF_CATEGORY(encoded_DF['day_Fri'])
            print(f":> TOTAL VISITS ON FRIDAY: {One_Binary.count(1)}")

        elif separations == 'day_Sat':
            CHECK_EVERY_TICK_OF_CATEGORY(encoded_DF['day_Sat'])
            print(f":> TOTAL VISITS ON SATURDAY: {One_Binary.count(1)}")

        elif separations == 'day_Sun':
            CHECK_EVERY_TICK_OF_CATEGORY(encoded_DF['day_Sun'])
            print(f":> TOTAL VISITS ON SUNDAY: {One_Binary.count(1)}")


# EXECUTE THE CATEGORY COUNTING FUNCTION FOR EACH ENCODED FEATURE
print("\n=>> ANALYZING ENCODED VALUES - SEX CATEGORY >>")
CHECK_CATEGORY_BINARY_VALUES('sex_Female')

print("=>> ANALYZING ENCODED VALUES - SMOKER CATEGORY >>")
CHECK_CATEGORY_BINARY_VALUES('smoker_No')

print("=>> ANALYZING ENCODED VALUES - TIME CATEGORY >>")
CHECK_CATEGORY_BINARY_VALUES('time_Dinner')


print("=>> ANALYZING ENCODED VALUES - DAY CATEGORY >>")
CHECK_CATEGORY_BINARY_VALUES('day_Thur')
CHECK_CATEGORY_BINARY_VALUES('day_Fri')
CHECK_CATEGORY_BINARY_VALUES('day_Sat')
CHECK_CATEGORY_BINARY_VALUES('day_Sun')
```
