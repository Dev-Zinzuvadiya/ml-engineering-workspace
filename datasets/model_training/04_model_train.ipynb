```python
print("=> METHOD 2: SCIKIT-LEANR ONEHOTENCODER (THE PROFESSIONAL WAY):\n")

oh_encoder = OneHotEncoder(sparse_output=False)

print("-> GRAB AND SELECT ONLY CATEGORICAL DATAs (COLUMNS):")
categorical_data = DataFram[['sex', 'smoker', 'day', 'time']]
categorical_data.head()
```

```python
print("-> APPLY ONE-HOT ENCODING:")
encoded_arr = oh_encoder.fit_transform(categorical_data)
display(encoded_arr)
```

```python
print("-> GET THE FEATURE NAMES AND CONVERT INTO DATAFRAME:\n")
features_names = oh_encoder.get_feature_names_out(['sex', 'smoker', 'day', 'time'])

encoded_df = pd.DataFrame(encoded_arr, columns=features_names)
print(encoded_df.head())
```

```python
store_OneValues = []
store_ZeroValues = []

def CHECK_EVERY_TICK_OF_CATEGORY(labels):
    store_OneValues.clear()
    store_ZeroValues.clear()

    for i in labels:
        if(i == 1):
            store_OneValues.append(i)
        else:
            store_ZeroValues.append(i)


def CHECK_CATEGORICAL_BINARY_VALUES(seprations):
    if seprations == 'sex_Female':
        CHECK_EVERY_TICK_OF_CATEGORY(encoded_df['sex_Female'])

        print(f"-> TOTAL MALES: {store_ZeroValues.count(0)}")
        print(f"-> TOTAL FEMALES: {store_OneValues.count(1)}\n")
        
    elif seprations == 'smoker_No':
        CHECK_EVERY_TICK_OF_CATEGORY(encoded_df['smoker_No'])

        print(f"-> TOTAL SMOKERS: {store_ZeroValues.count(0)}")
        print(f"-> TOTAL NON-SMOKERS: {store_OneValues.count(1)}\n")

    elif seprations == 'time_Dinner':
        CHECK_EVERY_TICK_OF_CATEGORY(encoded_df['time_Dinner'])

        print(f"-> TOTAL LUNCH VISITS: {store_ZeroValues.count(0)}")
        print(f"-> TOTAL DINNER VISITS: {store_OneValues.count(1)}\n")


    elif seprations == 'day_Thur' or seprations == 'day_Fri' or seprations == 'day_Sat' or seprations == 'day_Sun':

        if seprations == 'day_Thur':
            CHECK_EVERY_TICK_OF_CATEGORY(encoded_df['day_Thur'])
            print(f"-> TOTAL VISITS ON THURSHDAY: {store_OneValues.count(1)}")

        elif seprations == 'day_Fri':
            CHECK_EVERY_TICK_OF_CATEGORY(encoded_df['day_Fri'])
            print(f"-> TOTAL VISITS ON FRIDAY: {store_OneValues.count(1)}")

        elif seprations == 'day_Sat':
            CHECK_EVERY_TICK_OF_CATEGORY(encoded_df['day_Sat'])
            print(f"-> TOTAL VISITS ON SATURDAY: {store_OneValues.count(1)}")

        elif seprations == 'day_Sun':
            CHECK_EVERY_TICK_OF_CATEGORY(encoded_df['day_Sun'])
            print(f"-> TOTAL VISITS ON SUNDAY: {store_OneValues.count(1)}")


CHECK_CATEGORICAL_BINARY_VALUES('sex_Female')
CHECK_CATEGORICAL_BINARY_VALUES('smoker_No')
CHECK_CATEGORICAL_BINARY_VALUES('time_Dinner')

CHECK_CATEGORICAL_BINARY_VALUES('day_Thur')
CHECK_CATEGORICAL_BINARY_VALUES('day_Fri')
CHECK_CATEGORICAL_BINARY_VALUES('day_Sat')
CHECK_CATEGORICAL_BINARY_VALUES('day_Sun')
```
