```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer


# THIS EXAMPLE FOR SIMPLEIMPUTER..
data1 = pd.DataFrame(
    {
        "Age": [25, 30, np.nan, 35, 40, np.nan],
        "Income": [50000, 60000, 55000, np.nan, 65000, 70000],
    }
)

print("Original data with missing values:")
print(data1)

imputer = SimpleImputer(strategy="mean")
imputer.fit(data1)

print("\n\n--> Mean Calculated: ", imputer.statistics_)
data_filled = imputer.transform(data1)

print(f"\n\n{data_filled}")
```

```python
#  Different Strategies for Different Columns
player = pd.DataFrame(
    {
        "Player_Number": [1, 2, 3, 4, 5, 6],
        "Age": [86, 47, 38, np.nan, 33, np.nan],
        "Debt_Millions": [np.nan, 350, 600, 100, np.nan, 2000],
        "Game_Skill": ["High", "Medium", "High", np.nan, "Low", "Medium"],
    }
)

print("--> Squid Game Player Data (with missing info):")
print(player)

age_imputer = SimpleImputer(strategy="median")
debt_imputer = SimpleImputer(strategy="median")
skill_imputer = SimpleImputer(strategy="most_frequent")

player['Age'] = age_imputer.fit_transform(player[['Age']])
player['Debt_Millions'] = debt_imputer.fit_transform(player[['Debt_Millions']])
player['Game_Skill'] = skill_imputer.fit_transform(player[['Game_Skill']]).ravel()

print("\n\n--> After Front Man's investigation (imputation):")
print(player)


print("\n\n\n\n--> Another Example Different Strategies for Different Columns:")

# Create dataset with mixed types
data2 = pd.DataFrame(
    {
        "Age": [25, np.nan, 35, np.nan, 45],
        "Salary": [50000, 60000, np.nan, 80000, 90000],
        "Gender": ["M", "F", np.nan, "M", "F"],
        "City": ["NY", "LA", "CHI", np.nan, "NY"],
    }
)

print("Original messy data:")
print(data2)

num_imputer = SimpleImputer(strategy="median")
cat_imputer = SimpleImputer(strategy="most_frequent")
const_imputer = SimpleImputer(strategy="constant", fill_value=0)

data2["Age"] = num_imputer.fit_transform(data2[["Age"]])  # Median for age
data2["Salary"] = num_imputer.fit_transform(data2[["Salary"]])  # Median for salary
data2["Gender"] = cat_imputer.fit_transform(data2[["Gender"]]).ravel()  # Mode for gender
data2["City"] = cat_imputer.fit_transform(data2[["City"]]).ravel()  # Mode for city

print("\n\n--> After imputation:")
print(data2)
```
