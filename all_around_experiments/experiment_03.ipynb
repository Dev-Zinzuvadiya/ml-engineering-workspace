```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
CSVRowFiles = pd.read_csv("self_made.csv")
print(f">> TOTAL SHAPE OF DATASET : '{CSVRowFiles.shape}'")

print(">> HERE IS FEW DATASET RECORDS ↘")
display(CSVRowFiles)

print("\n>> INFO ABOUT DATASET COLUMNS ↘\n")
display(CSVRowFiles.info())
```

```python
print(">> NaN VALUEs TRANSFORM INTO BOOL DTYPE - AS A 'TRUE' ↘")
CSVRowFiles.isnull()
```

```python
print("\n>> COUNT OF NaN VALUES FROM DATASET ↘")
CSVRowFiles.isnull().sum()
```

```python
ViewsInPerc = CSVRowFiles.isnull().sum() / CSVRowFiles.shape[0] * 100

print("\n>> DISPLAY INTO 'PERCENTILE' EXCEPT NUMBERs ↘")
ViewsInPerc
```

```python
print(">> FILTER ONLY 'INT' AND 'FLOAT' DTYPE VALUEs ↘")
GrabDtype = CSVRowFiles.select_dtypes(include=['int64', 'float64'])
GrabDtype
```

```python
print(">> GIVE PERTICULAR ROWs WITH NaN VALUEs ↘")
display(GrabDtype[GrabDtype.isnull().any (axis=1)])
```

```python
print("\n>> DISPLAY NaN VALUEs COLUMNs ONLY ↘")
display(GrabDtype.isnull().sum())
```

```python
print(">> BRING COLUMN NAMES WHO HAVE NaN VALUES ↘")
[i for i in GrabDtype.columns if GrabDtype[i].isnull().sum() > 0]
```

```python
for i, j in enumerate(i for i in GrabDtype.columns if GrabDtype[i].isnull().sum() > 0):
    plt.figure(figsize=(15,15))
    plt.subplot(3,2,i+1)
    sns.set_theme()
    sns.histplot(GrabDtype[j], bins=20, kde=True)
```

```python
print(">> FILL NaN VALUES USING MEAN ↘")
display(GrabDtype.fillna(GrabDtype.mean()))
```

```python
print("\n>> DISPLAY AFTER FILL NaN VALUEs (COLUMNs ONLY) ↘")
display(GrabDtype.fillna(GrabDtype.mean()).isnull().sum())
```

```python

for i,j in enumerate(i for i in GrabDtype.columns if GrabDtype[i].isnull().sum() > 0):
    plt.figure(figsize=(15,15))
    plt.subplot(3,2,i+1)
    sns.histplot(GrabDtype[j], bins=20, kde=True, label='original')
    sns.histplot(GrabDtype.fillna(GrabDtype.mean())[j], bins=20, kde=True, label='mean')
    sns.set_theme()
    plt.legend()
    plt.show()
```
