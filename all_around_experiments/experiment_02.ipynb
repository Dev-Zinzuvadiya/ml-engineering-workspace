```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_fr1 = pd.read_csv('train.csv')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)

display(data_fr1.shape)
display(data_fr1.head(5))
```

```python
data_fr2 = data_fr1.isnull().sum() / data_fr1.shape[0] * 100
data_fr3 = data_fr2[data_fr2 > 20].keys()
data_fr4 = data_fr1.drop(columns = data_fr3)
data_fr5 = data_fr4.select_dtypes(include = ['int64', 'float64'])

data_fr6 = [i for i in data_fr5.columns if data_fr5[i].isnull().sum() > 0]
```

```python

for i,j in enumerate(data_fr6):
    plt.figure(figsize=(10,10))
    plt.subplot(2, 2, i + 1)
    sns.histplot(data_fr5[j], bins=20, kde=True)
    sns.set()
    plt.show()
```

```python
data_fr7 = data_fr5.fillna(data_fr5.mean())
display(data_fr7.isnull().sum().sum())
display(data_fr7)
```

```python

for i, j in enumerate(data_fr6):
    plt.figure(figsize=(10,10))
    plt.subplot(2,2,i+1)

    sns.histplot(data_fr5[j], bins=20, kde=True, label='original')
    sns.histplot(data_fr7[j], bins=20, kde=True, label='mean')
    
    # sns.distplot(data_fr5[j], bins=20, kde_kws={'linewidth':2, 'color':'blue'}, label='original')
    # sns.distplot(data_fr7[j], bins=20, kde_kws={'linewidth':2, 'color':'green'}, label='new')
    
    plt.legend()
    sns.set()
    plt.show()
```

```python
data_fr8 = data_fr5.fillna(data_fr5.median())
display(data_fr8.isnull().sum().sum())


for i,j in enumerate(data_fr6):
    plt.figure(figsize=(10,10))
    plt.subplot(2,2,i+1)
    
    sns.histplot(data_fr5[j], bins=20, kde=True, label='original')
    sns.histplot(data_fr7[j], bins=20, kde=True, label='mean')
    sns.histplot(data_fr8[j], bins=20, kde=True, label='median')
    
    plt.legend()
    sns.set()
    plt.show()
```

```python
data_fr9 = pd.concat([data_fr5[data_fr6], data_fr7[data_fr6], data_fr8[data_fr6]], axis=1)
data_fr9[data_fr9.isnull().any(axis=1)]
```
